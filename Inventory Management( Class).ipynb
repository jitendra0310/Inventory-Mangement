Created a class with name Inventory that takes in 5 arguments.
Three methods are defined in the Inventory class

```python
class Inventory(object):
    
   
    def __init__(self,p1,p2,p3,p4,p5):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4
        self.p5 = p5
    
    def layout(self):
        print('Current Inventory ')
        print('                   ')
        print('ID| Items  | Qty ')
        print('------------------')  
        print(f" {self.p1[0]}| {self.p1[1]}   |  {self.p1[2]}")
        print(f" {self.p2[0]}| {self.p2[1]}  |  {self.p2[2]}")
        print(f" {self.p3[0]}| {self.p3[1]}  |  {self.p3[2]}")
        print(f" {self.p4[0]}| {self.p4[1]}    |  {self.p4[2]}")
        print(f" {self.p5[0]}| {self.p5[1]} |  {self.p5[2]}")
        
            
    def inv_upd(self,x,y):
        p=[self.p1[2],self.p2[2],self.p3[2],self.p4[2],self.p5[2]]
        if x in range(1,6):
            if p[x-1] == 0:
                print('                                     ')
                print('Out of Stock, Please check again Later!')
            elif p[x-1] < y:
                print('                                     ')
                print('Requested quantity exceeds the available quality!')
            elif p[x-1] >= y:
                p[x - 1] -= y
                self.p1[2],self.p2[2],self.p3[2],self.p4[2],self.p5[2]=p
        else:
            print('Please enter a Valid Product ID!')
```

The Input to the Inventory class are given below.

```python
p1=[1,'Apple',10]
p2=[2,'Orange',12]
p3=[3,'Banana',6]
p4=[4,'Pear',5]
p5=[5,'Avacado',1]

total=Inventory(p1,p2,p3,p4,p5)
```

Driver class for the Inventory class, that takes the input and the methods of the inventory class

```python
print('                    Welcome to the Shipt Store                     ')
print('                                                                 ')
p=[p1[2],p2[2],p3[2],p4[2],p5[2]]
while True:
    total.layout()
    print('                                                                 ')
    x=int(input('Please enter the Product ID of the item : '))
    y=int(input('Please enter the Quantity of the item: '))
    total.inv_upd(x,y)
    if p[x-1] >=y:
        print('----- Your items have been ordered--------')
    print('                                                                 ')
    p=[p1[2],p2[2],p3[2],p4[2],p5[2]]
    if all(v==0 for v in p):
        total.layout()
        print('                                                                 ')
        print('Whole inventory is out of Stock!')
        break
```
